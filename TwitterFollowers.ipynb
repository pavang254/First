```
import tweepy
import csv
import time
import pandas as pd

consumer_key = 'Enter credentials here'
consumer_secret = 'Enter credentials here'
access_token = 'Enter credentials here'
access_token_secret = 'Enter credentials here'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```
followers_list = []   #create an empty list to store followers

for follower in tweepy.Cursor(api.followers, screen_name="FinlayRoyle").items():     #for each follower in the followers list we will iterate through cursor, the documentation of tweepy is in this format which we've written
  try:
      print('follower: ' + follower.screen_name)
      followers_list.append(follower.screen_name)   #now we need to add followers into our empty list, this can be done by append function
      time.sleep(3)
      #break

  except tweepy.TweepError as e:
    print(e.reason)
    #sleep(60*counter)
    #counter +=1
  except StopIteration:
    break
```

```
df = pd.DataFrame(followers_list)      #convert that list we made into a data frame by using pandas

df.shape
```

```
df.to_csv('FinlayRoyle_followers.csv')         #save into a csv file
```
